```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "liver_data.csv"
df = pd.read_csv(file_path)

# Ensure column names are stripped of leading/trailing spaces
df.columns = df.columns.str.strip()

# Convert categorical column ('Gender of the patient') into numeric values
df['Gender of the patient'] = df['Gender of the patient'].map({'Male': 1, 'Female': 0})

# Select only numerical columns for analysis
numerical_cols = df.select_dtypes(include=['number']).columns

# Set the style for all plots
sns.set_style("whitegrid")

# Create a figure with multiple subplots for detailed univariate analysis
fig, axes = plt.subplots(3, 2, figsize=(15, 12))

# 1. Univariate Analysis: Age Distribution
sns.histplot(df['Age of the patient'], bins=30, kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Count')

# 2. Univariate Analysis: Gender Distribution
sns.countplot(x=df['Gender of the patient'], ax=axes[0, 1], palette="pastel")
axes[0, 1].set_title('Gender Distribution')
axes[0, 1].set_xlabel('Gender (Male=1, Female=0)')
axes[0, 1].set_ylabel('Count')

# 3. Univariate Analysis: Total Bilirubin Distribution
sns.histplot(df['Total Bilirubin'], bins=30, kde=True, ax=axes[1, 0], color='purple')
axes[1, 0].set_title('Total Bilirubin Distribution')
axes[1, 0].set_xlabel('Total Bilirubin')
axes[1, 0].set_ylabel('Count')

# 4. Univariate Analysis: Direct Bilirubin Distribution
sns.histplot(df['Direct Bilirubin'], bins=30, kde=True, ax=axes[1, 1], color='green')
axes[1, 1].set_title('Direct Bilirubin Distribution')
axes[1, 1].set_xlabel('Direct Bilirubin')
axes[1, 1].set_ylabel('Count')

# 5. Univariate Analysis: SGPT (Alamine Aminotransferase) Distribution
sns.histplot(df['Sgpt Alamine Aminotransferase'], bins=30, kde=True, ax=axes[2, 0], color='red')
axes[2, 0].set_title('SGPT Distribution')
axes[2, 0].set_xlabel('SGPT')
axes[2, 0].set_ylabel('Count')

# 6. Univariate Analysis: SGOT (Aspartate Aminotransferase) Distribution
sns.histplot(df['Sgot Aspartate Aminotransferase'], bins=30, kde=True, ax=axes[2, 1], color='orange')
axes[2, 1].set_title('SGOT Distribution')
axes[2, 1].set_xlabel('SGOT')
axes[2, 1].set_ylabel('Count')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
# Create a figure with multiple subplots for detailed bivariate analysis
fig, axes = plt.subplots(3, 2, figsize=(15, 12))

# 1. Bivariate Analysis: Age vs Total Bilirubin
sns.scatterplot(x=df['Age of the patient'], y=df['Total Bilirubin'], ax=axes[0, 0], color='blue', alpha=0.5)
axes[0, 0].set_title('Age vs Total Bilirubin')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Total Bilirubin')

# 2. Bivariate Analysis: Gender vs Direct Bilirubin (Boxplot)
sns.boxplot(x=df['Gender of the patient'], y=df['Direct Bilirubin'], ax=axes[0, 1], palette="pastel")
axes[0, 1].set_title('Gender vs Direct Bilirubin')
axes[0, 1].set_xlabel('Gender (Male=1, Female=0)')
axes[0, 1].set_ylabel('Direct Bilirubin')

# 3. Bivariate Analysis: SGPT vs SGOT (Liver Enzyme Relationship)
sns.scatterplot(x=df['Sgpt Alamine Aminotransferase'], y=df['Sgot Aspartate Aminotransferase'], ax=axes[1, 0], color='purple', alpha=0.5)
axes[1, 0].set_title('SGPT vs SGOT')
axes[1, 0].set_xlabel('SGPT')
axes[1, 0].set_ylabel('SGOT')

# 4. Bivariate Analysis: Total Proteins vs Albumin
sns.scatterplot(x=df['Total Protiens'], y=df['ALB Albumin'], ax=axes[1, 1], color='green', alpha=0.5)
axes[1, 1].set_title('Total Proteins vs Albumin')
axes[1, 1].set_xlabel('Total Proteins')
axes[1, 1].set_ylabel('Albumin')

# 5. Bivariate Analysis: Age vs SGOT
sns.scatterplot(x=df['Age of the patient'], y=df['Sgot Aspartate Aminotransferase'], ax=axes[2, 0], color='red', alpha=0.5)
axes[2, 0].set_title('Age vs SGOT')
axes[2, 0].set_xlabel('Age')
axes[2, 0].set_ylabel('SGOT')

# 6. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the liver dataset
liver_data_path = "liver_data.csv"  # Update path as necessary
liver_data = pd.read_csv(liver_data_path)

# Clean column names by stripping spaces
liver_data.columns = liver_data.columns.str.strip()

# Handling missing values using forward-fill and back-fill
liver_data_cleaned = liver_data.ffill().bfill()

# Selecting key biomarkers for time series analysis
biomarkers = [
    "Total Bilirubin",
    "Direct Bilirubin",
    "Alkphos Alkaline Phosphotase",
    "Sgpt Alamine Aminotransferase",
    "Sgot Aspartate Aminotransferase",
    "Total Protiens",
    "ALB Albumin",
    "A/G Ratio Albumin and Globulin Ratio"
]

# Creating a time index assuming sequential data collection
liver_data_cleaned["Time Index"] = range(len(liver_data_cleaned))
liver_data_cleaned.set_index("Time Index", inplace=True)

# Plotting time series trends for the selected biomarkers
for biomarker in biomarkers:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=liver_data_cleaned[biomarker])
    plt.title(f"Time Series of {biomarker}")
    plt.xlabel("Time Index")
    plt.ylabel(biomarker)
    plt.show()

# Applying a cumulative sum transformation to simulate a random walk
liver_data_cumsum = liver_data_cleaned[biomarkers].cumsum()

# Plotting the cumulative sum to observe long-term trends
for biomarker in biomarkers:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=liver_data_cumsum[biomarker])
    plt.title(f"Cumulative Sum (Random Walk) of {biomarker}")
    plt.xlabel("Time Index")
    plt.ylabel(biomarker)
    plt.show()

# Applying a moving average (window=50) to smooth trends
window_size = 50

for biomarker in biomarkers:
    plt.figure(figsize=(12, 6))
    plt.plot(liver_data_cleaned[biomarker], label="Original", alpha=0.5)
    plt.plot(liver_data_cleaned[biomarker].rolling(window=window_size).mean(), label=f"Moving Average ({window_size})", color="red")
    plt.title(f"Smoothed Time Series of {biomarker}")
    plt.xlabel("Time Index")
    plt.ylabel(biomarker)
    plt.legend()
    plt.show()
```
